#What is data cleaning, and why is it important in data analysis?
 (A)What are the potential consequences of analyzing unclean or messy data?
 (B)Explain the common steps involved in cleaning and organizing data.

Ans1)Data cleaning (also called data cleansing or data preprocessing) is the process of detecting and correcting (or removing) inaccurate, incomplete, inconsistent, or irrelevant data from a dataset.
It ensures that the data is accurate, reliable, and ready for analysis.

In simple terms — it’s like polishing raw data so that analysis gives meaningful and trustworthy results.
Clean data is the foundation of accurate analysis.
If your data is wrong, your conclusions will also be wrong — no matter how advanced your analysis or models are.

Importance:

✅ Ensures accuracy of insights and results

✅ Improves data consistency across sources

✅ Enhances decision-making quality

✅ Reduces processing errors in analysis or machine learning models

✅ Saves time and cost in the long run

(A)Potential Consequences of Analyzing Unclean or Messy Data

If unclean data is analyzed, the outcomes can be misleading or even harmful.

Consequences:

❌ Incorrect insights or conclusions — leading to poor business or policy decisions

❌ Skewed analysis or bias — due to missing or duplicate values

❌ Reduced accuracy of machine learning models

❌ Loss of credibility — analysts or organizations appear unreliable

❌ Wasted time and resources — because wrong results may need rework.

#Q3) Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

Ans3) Text functions are used in data cleaning to standardize, extract, and combine textual data, ensuring consistency and usability for analysis.
TRIM: Removes leading, trailing, and often excess internal spaces from text, which is crucial for standardizing entries and ensuring accurate comparisons.
LEFT/RIGHT: Extracts a specified number of characters from the beginning (LEFT) or end (RIGHT) of a text string, useful for isolating specific codes or identifiers.
MID: Extracts a substring from a text string starting at a specified position and for a specified length, allowing for extraction of data embedded within a larger string.
CONCAT: Combines multiple text strings into a single string, useful for merging separate data fields (e.g., first and last names) into a complete entry.

Ans8) (a) Difference between TODAY() and NOW()
Answer: TODAY() returns only the current date, while NOW() returns both the current date and time.
Explanation:
=TODAY() returns a date value that updates whenever the workbook is opened or a change is made that forces a recalculation.
=NOW() returns both the date and time value, which also updates dynamically.
Use =TODAY() when you only need to calculate based on the current date, such as determining age or a deadline date.
Use =NOW() when you need a precise timestamp, such as logging the exact time of a transaction or entry.

Explanation:
